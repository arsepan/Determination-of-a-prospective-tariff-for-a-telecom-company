```python
# Сформируем общий DF пользователя по использованию тарифа

# DF internet
# Создадим сводную таблицу, группируя значения по пользователям и месяцам, найдём сумму использованных mb  в кажом месяце
pivot_internet = internet.pivot_table(index=['user_id', 'moment_of_use'], values='mb_used', aggfunc='sum')
# Создадим новый столбце на основе суммы МБ - общая сумма ГБ каждого пользователя по месяцам и удалим старый ненужный нам столбец
pivot_internet['gb_used'] = (pivot_internet['mb_used'] / 1024)
pivot_internet['gb_used'] = np.ceil(pivot_internet['gb_used'])
del pivot_internet['mb_used']
# Выведем полученный результат
pivot_internet.head(10)
```

```python
# DF calls
# Создадим сводную таблицу, группируя значения по пользователям и месяцам, найдём сумму использованных минут и количество звонков в кажом месяце
pivot_calls = calls.pivot_table(index=['user_id', 'moment_of_use'], values='duration', aggfunc={'sum', 'count'}) 
pivot_calls.rename(columns={'count' : 'calls_count', 'sum' : 'calls_sum'}, inplace = True)
pivot_calls.head(10)
```

```python
# DF messages

# Создадим своздую таблицу, группируя значения по пользователям и месяцам, найдём бщее кол-во отправленных СМС каждого пользователя за месяц
pivot_messages = messages.pivot_table(index=['user_id', 'moment_of_use'], aggfunc={'moment_of_use' : 'count'})
pivot_messages.rename(columns={'moment_of_use' : 'messages_count'}, inplace=True)
pivot_messages
```

```python
# Соеденим все полученные таблицы
all_pivots = pivot_calls.join(pivot_messages, how='outer')
all_pivots = all_pivots.join(pivot_internet, how='outer')
all_pivots.reset_index(level=['moment_of_use', 'user_id'], inplace=True)
all_pivots
```

```python
users_summary = users.merge(all_pivots, on='user_id', how='outer')
users_summary.head(10)
```

```python
# Проверим пропуски, которые могли образоваться ввиду объединения таблиц
users_summary.isna().sum()
```

```python
# Удалим все NaN значения в столбце month_of_call, тк мы не можем заменить их на 0, тем более их всего лишь 8
users_summary = users_summary.dropna(subset=['month_number'])
# Заменим все NaN значения нулями (Кроме churn_date), также удилим ненужным нам столбце churn_date
users_summary = users_summary.fillna(0)
display(users_summary.isna().sum())
users_summary
```

```python
# Напишем функцию, которая рассчитает выручку с одного клиента помесячно
def revenue(row):
    summ = 0 
    if row['tariff'] == 'ultra':
        summ += 1950
        if row['gb_used'] > (row['mb_per_month_included'] / 1024):
            summ += (row['gb_used'] - (row['mb_per_month_included'] / 1024)) * row['rub_per_gb']
        if row['calls_sum'] > row['minutes_included']:
            summ += (row['calls_sum'] - row['minutes_included']) * row['rub_per_minute']
        if row['messages_count'] > row['messages_included']:
            summ += (row['messages_count'] - row['messages_included']) * row['rub_per_message']
    else:
        summ += 550
        if row['gb_used'] > (row['mb_per_month_included'] / 1024):
            summ += (row['gb_used'] - (row['mb_per_month_included'] / 1024)) * row['rub_per_gb']
        if row['calls_sum'] > row['minutes_included']:
            summ += (row['calls_sum'] - row['minutes_included']) * row['rub_per_minute']
        if row['messages_count'] > row['messages_included']:
            summ += (row['messages_count'] - row['messages_included']) * row['rub_per_message'] 
    return summ
```

```python
# Создадим новый столце методом apply
users_summary['total_revenues'] = users_summary.apply(revenue, axis=1)
```

## Проверка гипотез

Давайте проверим первую гипотезу: средняя выручка пользователей тарифов «Ультра» и «Смарт» различаются.

1. Нулевая гипотеза H0 - средняя выручка пользователей тарифов «Ультра» и «Смарт» равны (всегда используется знак равенства)

2. Альтернативная гипотеза H1 - средняя выручка пользователей тарифов «Ультра» и «Смарт» различаются

```python
smart_revenues = users_summary[users_summary['tariff'] == 'smart']['total_revenues']
ultra_revenues = users_summary[users_summary['tariff'] == 'ultra']['total_revenues']

alpha = .05

results = st.ttest_ind(smart_revenues, ultra_revenues)

print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу") 
```

### Вывод "Первая гипотеза"

Мы можем увидеть, что p-значение намного меньше, чем наш критический уровень статистической значимости в 5%, следовательно, - мы отвергаем нулевую гипотезу, средняя выручка пользователей тарифов «Ультра» и «Смарт» сильно различаются.

Давайте проверим вторую гипотезу: средняя выручка пользователей из Москвы отличается от выручки пользователей из других регионов.
    
1. Нулевая гипотеза (H0) - средняя выручка пользователей из Москвы равна выручки пользователей из других регионов
2. Альтернативная гипотеза (H1) - средняя выручка пользователей из Москвы отличается от выручки пользователей из других регионов

В данной ситуации мы должны использлвать метод st.ttest_1samp, а не st.ttest_ind. У них есть некоторые различия:
1. 1samp используется при проверке двусторонней и односторонней гипотезы, он сравнивает генеральную совокупность (в нашем случае выборку) с определённым числом (в нашем случае средним показателем)
2. ind используется для сравнения двух средних значений у независимых между собой выборок (в отличии от метода st.ttest_rel, который сравнивает парные выборки)

```python
moscow_users_revenues = users_summary[users_summary['city'] == 'Москва']['total_revenues']
nmoscow_users_revenues = users_summary[users_summary['city'] != 'Москва']['total_revenues']

alpha = .05

results = st.ttest_ind(moscow_users_revenues, nmoscow_users_revenues)

print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу") 
```

С точки зрения статистики **p-value** - величина, которая используется при проверке статистических гипотез. Строго говоря, это вероятность ошибки отклонении заведомо правильной гипотезы (ошибка первого рода). Мы никогда не можем быть уверенными на 100%, что та или иная гипотеза верная. Наоборот, никакие эксперементально полученные данные не подтвердят её. Данные могут лишь не противоречить ей, либо говорить о том, что она маловероятна, однако ни в каком из двух случаев нельзя говорить, что гипотеза доказана. 

Приведу пример (p-значение: 0.5440152911670224), это можно интерпретировать следубщим образом: хоть средние чеки москвичей и не москвичей неравны, с вероятностью 54% случайным образом такое или большее различие можно получить случайно. Ведь это выборка, мы не можем отталкиваться от неё, как от генеральной совокупности. 

### Вывод "Вторая гипотеза"

Мы можем увидеть, что p-значение больше (47%), чем наш критический уровень статистической значимости в 5%, следовательно, - мы не можем отвергнуть данную гипотезу - средняя выручка пользователей из Москвы не отличается от выручки пользователей из других регионов.

```python
# Приведены два датасета: сумма покупок, совершённых за месяц посетителями ...

sample_1 = [3071, 3636, 3454, 3151, 2185, 3259, 1727, 2263, 2015,
2582, 4815, 633, 3186, 887, 2028, 3589, 2564, 1422, 1785,
3180, 1770, 2716, 2546, 1848, 4644, 3134, 475, 2686,
1838, 3352]
sample_2 = [1211, 1228, 2157, 3699, 600, 1898, 1688, 1420, 5048, 3007,
509, 3777, 5583, 3949, 121, 1674, 4300, 1338, 3066,
3562, 1010, 2311, 462, 863, 2021, 528, 1849, 255,
1740, 2596]
alpha = .05 # критический уровень статистической значимости
# если p-value окажется меньше него - отвергнем гипотезу
results = st.ttest_ind(
sample_1,
sample_2)
print('p-значение:', results.pvalue)
if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу")
else:

    print("Не получилось отвергнуть нулевую гипотезу")
```

```python
# Давайте проверим следующую гипотезу: "Большинство ушедших пользователей тарифа smart страдали переплатой"
# H0 - Медианное значение по всем месяцам у каждого пользователя равно тарифной платы (550)
# H1 - Медианное значение по всем месяцам у каждого пользователя будет больше тарифной платы (550)
smart_churn_list = list(users.loc[(users['tariff'] == 'smart') & (users['churn_month'].notna()), 'user_id'].reset_index(drop=True))
smart_churn = users_summary.query('user_id in @smart_churn_list').groupby('user_id')['total_revenues'].median()

smart_value = 550

alpha = .01 # уровень статистической значимости

results = st.ttest_1samp(
    smart_churn, 
    smart_value)

# тест односторонний: p-value будет в два раза меньше
print('p-значение:', results.pvalue / 2)

# отвергаем гипотезу только тогда, когда выборочное среднее значимо меньше предполагаемого значения
if (results.pvalue / 2 < alpha) and (smart_churn.median() > smart_value):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не отвергаем нулевую гипотезу") 
```
